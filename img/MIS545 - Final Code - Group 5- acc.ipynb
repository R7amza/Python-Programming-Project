```python
def PlotBarCharts(inpData, colsToPlot):
    
    import matplotlib.pyplot as plt
    
    # Generating multiple subplots
    fig, subPlot=plt.subplots(nrows=1, ncols=len(colsToPlot), figsize=(40,6))
    fig.suptitle('Bar charts of: '+ str(colsToPlot))

    for colName, plotNumber in zip(colsToPlot, range(len(colsToPlot))):
        inpData.groupby(colName).size().plot(kind='bar',ax=subPlot[plotNumber])
```

```python
PlotBarCharts(inpData=acc, colsToPlot=['gender', 'age', 'edu'])
PlotBarCharts(inpData=acc, colsToPlot=[ 'major', 'internet_usage', 'skills_level'])
PlotBarCharts(inpData=acc, colsToPlot=['used_device', 'connectivity_type', 'usage_purpose'])
PlotBarCharts(inpData=acc, colsToPlot=['computer_os', 'mobile_os', 'security_tools'])
PlotBarCharts(inpData=acc, colsToPlot=[ 'web_legitimacy', 'pass_weakness','danger_awarness'])
PlotBarCharts(inpData=acc, colsToPlot=[ 'password_change', 'public_wifi', 'link_click'])
PlotBarCharts(inpData=acc, colsToPlot=['phishing_email', 'malware_infection'])
```

<p><strong><span style="font-family: Calibri, sans-serif; font-size: 28px;">Statistical Feature Selection</span></strong><span style="font-family: Calibri, sans-serif; font-size: 28px;"><strong>&nbsp;</strong></span></p>

<p><span style="font-family: Calibri, sans-serif; font-size: 18px;">(Categorical Vs Categorical) using Chi-Square Test</span></p>

```python
# Writing a function to find the correlation of all categorical variables with the Target variable
def FunctionChisq(inpData, TargetVariable, CategoricalVariablesList):
    from scipy.stats import chi2_contingency
    
    # Creating an empty list of final selected predictors
    SelectedPredictors=[]

    for predictor in CategoricalVariablesList:
        CrossTabResult=pd.crosstab(index=inpData[TargetVariable], columns=inpData[predictor])
        ChiSqResult = chi2_contingency(CrossTabResult)
        
        # If the ChiSq P-Value is <0.05, that means we reject H0
        if (ChiSqResult[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])        
            
    return(SelectedPredictors)
```

```python
CategoricalVariables=['gender', 'age', 'edu', 'skills_level', 'computer_os',
       'mobile_os', 'web_legitimacy', 'pass_weakness', 'danger_awarness',
       'password_change', 'public_wifi', 'link_click', 'phishing_email',
       'malware_infection']

# Calling the function
FunctionChisq(inpData=acc, 
              TargetVariable='been_victim',
              CategoricalVariablesList= CategoricalVariables)
```

<p><span style="font-family: Calibri, sans-serif; font-size: 18px;">To statistically identify the significant related variables with the targert variable, a Chi-Square test has been conducted. As a result, this set of variables ['edu', 'web_legitimacy', 'public_wifi', 'phishing_email', 'malware_infection'] are the most significant to build the classification model.</span></p>

<p><span style="font-family: Calibri, sans-serif;"><strong><span style="font-size: 28px;">Set the Model Evaluation Metrics</span></strong></span></p>

```python
def model_eval(matrix,model_name):
            tn = matrix.iloc[0,0]
            tp = matrix.iloc[1,1]
            fn = matrix.iloc[1,0]
            fp = matrix.iloc[0,1]
            tap = fn+tp
            tan = tn+fp
            tpn = tn+fn
            tpp = fp+tp
            precision = tp/tpp
            recall = tp/tap
            total = tn+tp+fn+fp
            data = [
                    round((tp+tn)/total,4),
                    round(1-((tp+tn)/total),4),
                    round(tp/tap,4),
                    round(tn/tan,4),
                    round(precision,4),
                    round(2 * (precision * recall) / (precision + recall),4),
                    round(5 * (precision * recall) / ((4 * precision) + recall),4),
                    round(1.25 * (precision * recall) / ((.25 * precision) + recall),4),
                    ]
            return(
                    pd.DataFrame(data, columns=[model_name],
                    index=['Accuracy','Error Rate','Sensitivity','Specificity','Precision','F1','F2','F3'])
)
```

```python
# Apply final model, we select Logistic Regression
from sklearn.linear_model import LogisticRegression

# Using the Logistic Regression algorithm with final hyperparamters
clf = LogisticRegression(C=1,penalty='l2', solver='newton-cg')

# Training the model on the available data and we name it FinalLRM
FinalLRM=clf.fit(X,y)
```

```python
# Importing cross validation function from sklearn
from sklearn.model_selection import cross_val_score

# Running 10-Fold Cross validation on a given algorithm
# Passing full data X and y because the K-fold will split the data and automatically choose train/test
# this step will help to see the impact of running the model multiple time as this what will be done in the production phase

Accuracy_Values=cross_val_score(FinalLRM, X , y, cv=10, scoring='f1_weighted')
print('\nAccuracy values for 10-fold Cross Validation:\n',Accuracy_Values)
print('\nFinal Average Accuracy of the model:', round(Accuracy_Values.mean(),2))
```
